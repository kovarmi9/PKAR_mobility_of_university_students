```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import LineString
from shapely.geometry import Point
import matplotlib.pyplot as plt
from pyproj import Transformer
```

```python
# 1. Načti data
df = pd.read_excel('../../data/czech/data_czech.xlsx')

# 2. Ulož název krajů ze sloupce a nastav index
df = df.rename(columns={'University ↓ / Student →': 'origin'})
df = df.set_index('origin')

# 3. Očisti názvy (pro jistotu, kvůli shodě)
df.index = df.index.str.strip()
df.columns = df.columns.str.strip()

# 4. Pro každou krajskou kombinaci: pokud je řádek == sloupec → nastav 0
for region in df.index:
    if region in df.columns:
        df.at[region, region] = 0

# 5. Výstup pro kontrolu
print(df)
```

```python
# Příchozí studenti do každého kraje
prichozi = df.sum(axis=1).rename('prichozi')

print(prichozi)
```

```python
# Odchozí studenti z každého kraje
odchozi = df.sum(axis=0).rename('odchozi')

print(odchozi)
```

```python
# Load spatial regions
gdf = gpd.read_file('../../data/czech/regions.gpkg')

# Keep only necessary columns and compute centroids
gdf = gdf[['nazev', 'geometry']]

# Compute default centroids
gdf['centroid'] = gdf.representative_point()

# Prepare transformation from WGS84 to the CRS of gdf
transformer = Transformer.from_crs("EPSG:4326", gdf.crs, always_xy=True)

# Define manual centroids in WGS84 and transform them
x_stc, y_stc = transformer.transform(14.20, 49.80)     # Středočeský kraj – Kolín
x_olk, y_olk = transformer.transform(17.15, 49.65)     # Olomoucký kraj – lehce na JV
x_prh, y_prh = transformer.transform(14.50, 50.10)
x_par, y_par = transformer.transform(16.45, 49.85)

# Dictionary with manual overrides
manual_centroids = {
    'Středočeský kraj': Point(x_stc, y_stc),
    'Olomoucký kraj': Point(x_olk, y_olk),
        'Hlavní město Praha': Point(x_prh, y_prh),
    'Pardubický kraj': Point(x_par, y_par)
}

# Apply manual centroids if applicable
gdf['centroid'] = gdf.apply(
    lambda row: manual_centroids.get(row['nazev'], row['centroid']),
    axis=1
)

# Show the result
gdf.head()
```

```python
# Převod na DataFrame a zajištění správného indexu
df_prichozi = prichozi.reset_index().set_index('origin')
df_odchozi = odchozi.reset_index().set_index('index')
df_prichozi
```

```python
# Nejdřív merge příchozích
gdf = gdf.merge(df_prichozi, left_on='nazev', right_index=True, how='left')

# Potom merge odchozích
gdf = gdf.merge(df_odchozi, left_on='nazev', right_index=True, how='left')

# Vyplníme případné NaNy nulou
gdf[['prichozi', 'odchozi']] = gdf[['prichozi', 'odchozi']].fillna(0).astype(int)
gdf
```

```python
gdf['celkem'] = gdf['prichozi'] + gdf['odchozi']
gdf
```

```python
# 1. Set up the figure and axes
fig, ax = plt.subplots(1, 1, figsize=(12, 10))

# 2. Plot regional polygons (e.g., administrative boundaries)
gdf.plot(ax=ax, color='lightgrey', edgecolor='white')

# 3. Plot centroids for total number of students (in red)
gdf.set_geometry('centroid').plot(
    ax=ax,
    markersize=gdf['celkem'] / 100,  # Scale by total number of students
    color='red',                     # Red color for total students
    alpha=0.7,                       # Transparency
    label='Total number of students'
)

# 4. Plot centroids for incoming students (in green)
gdf.set_geometry('centroid').plot(
    ax=ax,
    markersize=gdf['prichozi'] / 100,  # Scale by number of incoming students
    color='limegreen',                 # Green color for incoming students
    alpha=0.7,                         # Transparency
    label='Incoming students'
)

# 5. Add title, legend, and final layout adjustments
ax.set_title("Ratio of incoming (green) to total (red) students by region", fontsize=16)
ax.axis('off')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Assign the 'centroid' column as the new geometry
gdf['geometry'] = gdf['centroid']

# Drop the original 'centroid' column
gdf = gdf.drop(columns='centroid')

# Set the correct coordinate reference system – EPSG:3035 (ETRS89 / LAEA Europe)
gdf = gdf.set_crs(epsg=3035)

# Export to GeoPackage
gdf.to_file('../../data/czech/czech_centriods.gpkg', driver='GPKG', layer='centroid_data')

print("✅ GeoPackage successfully exported as 'data/czech/czech_centriods.gpkg' in EPSG:3035.")
```
